# Estimated revenue

```python
all(df['price'].apply(lambda x: x.startswith('$')))
```

```python
def intify(dollarstr):
    return int(float(''.join(dollarstr.split('$')[1].split(','))))
print(intify('$12,245.00'))
print(intify('$124.00'))
```

```python
df['price_usd'] = df['price'].apply(intify)
```

```python
avail_tups
```

```python
for day, win in avail_tups:
    colname = 'revenue_{}'.format(day)

    df[colname] = df['price_usd'] * (day - df[win])
```

```python
revenue_periods = ['revenue_{}'.format(day) for day in days]
df[['price']+revenue_periods+avail_windows]
```

```python
hosts_revenue = df.groupby('multilister')
```

```python
agg_rev={revenue_period: sum for revenue_period in revenue_periods}
```

```python
multilist_revenues = hosts_revenue.aggregate(agg_rev).T
```

```python
multilist_revenues['total'] = multilist_revenues.sum(axis=1)
multilist_revenues['multilist%'] = multilist_revenues[True] / multilist_revenues['total']
```
